## Importing library

```python
import urllib2
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
```

## Specify the url

```python
national = "http://www.necliberia.org/results2011/national.html"
```

## Query the website and return the html to the variable page

```python
page = urllib2.urlopen(national)
```

## Pare the html in the 'page' variable and store it in BeautifulSoup format

```python
soup = BeautifulSoup(page)
```

```python
print soup.prettify()
```

```python
print soup.title.getText()
```

## Find all tables

```python
all_result = soup.findAll(class_="res")
```

```python
pres_vp_table1 = all_result[0].find("table")
pres_vp_table2 = all_result[1].find("table")
```

## Generate lists

```python
candidates = []
votes = []
percents = []
```

## Get data from first table

```python
for row in pres_vp_table1.findAll("tr"):
    cells = row.findAll('td')
    n = cells[0].get_text()
    b = cells[1].get_text()
    percent = cells[2].get_text()
    candidates.append(n)
    votes.append(b)
    percents.append(percent)
```

## Make a dataframe

```python
result1 = pd.DataFrame({'Candidate':candidates,'votes':votes,'percent':percents})
```
